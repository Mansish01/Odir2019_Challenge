```python

all_labels = df["Left-Diagnostic Keywords"].tolist() + df["Right-Diagnostic Keywords"].tolist()
print(all_labels)

# get unique diagnostic keywords
keywords = [ keyword  for keywords in all_labels for keyword in keywords.split('，')]
unique_keywords = set(keywords)    
    

print(unique_keywords)
print(len(unique_keywords), len(all_labels))
```

```python
# create a mapping from keywords to class labels
class_labels = ['N','D','G','C','A','H','M','O']
keyword_label_mapping  = {
    'normal':'N',
    'retinopathy':'D',
    'glaucoma':'G',
    'cataract':'C',
    'macular degeneration':'A',
    'hypertensive':'H',
    'myopia':'M',
}
non_decisive_labels = ["lens dust", "optic disk photographically invisible", "low image quality", "image offset"]
```

```python
def generate_label(diagonostic_label):
     keywords = [ keyword  for keyword in diagonostic_label.split('，')]
     print(keywords)
     for k in keywords:
        for label in keyword_label_mapping.keys():
            if label in k:
               
                       return keyword_label_mapping[label]

     
     for k in keywords:
         
        for label in keyword_label_mapping.keys():
            if label not in k:   
                 return 'O'  
                           
         

    
```

```python
generate_label('wet age-related macular degeneration'), generate_label('severe nonproliferative retinopathy，hypertensive retinopathy'), generate_label('epiretinal membrane')

```

```python
generate_label('normal fundus , lens dust'),generate_label('lens dust，drusen，normal fundus'), generate_label('moderate non proliferative retinopathy')
```

```python
df['Left_Label'] = df['Left-Diagnostic Keywords'].apply(generate_label)
df['Right_Label'] = df['Right-Diagnostic Keywords'].apply(generate_label)
```

```python
SEED = 42
train_id , test_id =train_test_split(id, test_size = 0.2, random_state= SEED)

train_left_fundus = df[df['ID'].isin(train_id)]['Left-Fundus']
# print(train_left_fundus)
train_left_label = df[df['ID'].isin(train_id)]['Left_Label']
train_left_keywords = df[df['ID'].isin(train_id)]['Left-Diagnostic Keywords']


train_right_fundus = df[df['ID'].isin(train_id)]['Right-Fundus']
train_right_label = df[df['ID'].isin(train_id)]['Right_Label']
train_right_keywords = df[df['ID'].isin(train_id)]['Right-Diagnostic Keywords']

test_left_fundus = df[df['ID'].isin(test_id)]['Left-Fundus']
test_left_label = df[df['ID'].isin(test_id)]['Left_Label']
test_left_keywords = df[df['ID'].isin(test_id)]['Left-Diagnostic Keywords']


test_right_fundus = df[df['ID'].isin(test_id)]['Right-Fundus']
test_right_label = df[df['ID'].isin(test_id)]['Right_Label']
test_right_keywords = df[df['ID'].isin(test_id)]['Right-Diagnostic Keywords']



```

```python
train_image_fundus = pd.concat([train_left_fundus, train_right_fundus], axis=0)
train_label = pd.concat([train_left_label, train_right_label], axis = 0)
train_keywords = pd.concat([train_left_keywords, train_right_keywords], axis= 0)

test_image_fundus = pd.concat([test_left_fundus, test_right_fundus], axis=0)
test_label = pd.concat([test_left_label, test_right_label], axis = 0)
test_keywords = pd.concat([test_left_keywords, test_right_keywords], axis= 0)
```
